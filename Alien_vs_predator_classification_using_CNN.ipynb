```python
# Importing necessary libraries
import numpy as np
import os
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
```

```python
target = [] # to store label
images = [] # to store resized images
flattened_data = [] # to store flat data
test_images = [] # to store resized images for test data

train_data_path = "C:/Users/Rahul/ALIEN_vs_PREDATOR/data/train"
test_data_path = "C:/Users/Rahul/ALIEN_vs_PREDATOR/data/validation"
categories = ["alien", "predator"]

# for loop for train data
for cat in categories:
    category_label = categories.index(cat)
    item_path = os.path.join(train_data_path, cat)
    for img in os.listdir(item_path):
        image_path = os.path.join(item_path, img)
        image_array = imread(image_path)
        #print(image_array.shape)
        #plt.imshow(image_array)
        #break
        resized_image = resize(image_array, (150, 150, 3))
        target.append(category_label)
        images.append(resized_image)
        flattened_data.append(resized_image.flatten())

# for loop for test data
for cat in categories:
    test_item_path = os.path.join(test_data_path, cat)
    for img in os.listdir(test_item_path):
        test_image_path = os.path.join(test_item_path, img)
        test_image_array = imread(test_image_path)
        test_resized_image = resize(test_image_array, (150, 150, 3))
        test_images.append(test_resized_image)
        
target = np.array(target)
images = np.array(images)
flattened_data = np.array(flattened_data)
test_images = np.array(test_images)
```

```python
# Defining categories of label
from keras.utils.np_utils import to_categorical
target = to_categorical(target, num_classes=2)
target.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(images, target, test_size=0.3, random_state=2)
```

```python
plt.imshow(np.array(X_train[5]))
```

```python
# Importing necessary libraries of keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, BatchNormalization
from keras.optimizers import RMSprop
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ReduceLROnPlateau
```

```python
# Using convolution layer
model = Sequential()
model.add(Conv2D(filters=64, kernel_size=(5,5), padding='Same', activation='relu', input_shape=(225, 300,1)))
model.add(BatchNormalization())
```

```python
model.add(Conv2D(filters=64, kernel_size=(5,5), padding='Same', activation='relu'))
model.add(BatchNormalization())
```

```python
# Pooling layer acts as downsampling filter
#  It looks at the 2 neighboring pixels and picks the maximal value.
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))
```

```python
model.add(Conv2D(filters=64, kernel_size=(3,3), padding='Same', activation='relu'))
model.add(BatchNormalization())
```

```python
model.add(Conv2D(filters=64, kernel_size=(3,3), padding='Same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))
# Dropout is a regularization method, where a proportion 
# of nodes in the layer are randomly ignored 
# (setting their weights to zero) for each training sample)
```

```python
model.add(Conv2D(filters=64, kernel_size=(3,3), padding='Same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.25))

model.add(Dense(2, activation='softmax'))
model.summary()
```

```python
optimizer = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
```

```python
model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
# Set annealing method of learning rate
#  to make the optimizer converge faster and 
# closest to the global minimum of the loss function
learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', patience=3, verbose=1, factor=0.5, min_lr=0.0)
```

```python
# Increase epochs for better accuracy, here I'm posting for epoch = 1 
# because It will take time to run for more number epochs
epochs=3
batch_size= 30
```

```python
history = model.fit(X_train, Y_train, batch_size= batch_size, epochs = epochs, validation_data=(X_val, Y_val), verbose=2)
```
